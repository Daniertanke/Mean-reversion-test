# Testeo Estrategia de Reversión a la Media 

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
```

Voy a importar los precios de cierre de 4 índices primero para ver el desempeño desde 2005 de la estrategia

```python
SP500_data = wb.DataReader('^GSPC','yahoo',start='2005-1-1')   #DataFrame SP500
SP500_data['price'] = SP500_data['Adj Close']

NASDAQ_data = wb.DataReader('^IXIC','yahoo',start='2005-1-1')   #DataFrame Nasdaq
NASDAQ_data['price'] = NASDAQ_data['Adj Close']

DAX_data = wb.DataReader('^GDAXI','yahoo',start='2005-1-1')     #DataFrame Dax
DAX_data['price'] = DAX_data['Adj Close']

IBEX_data = wb.DataReader('^IBEX','yahoo',start='2005-1-1')     #DataFrame Ibex
IBEX_data['price'] = IBEX_data['Adj Close']
```

Vamos a crear un modelo que si el precio está a una distancia equivalente a la desviación estándar respecto del precio ejecute una operación. Si está por encima la operación será de venta y si está por debajo de compra.

```python
SMA = 25 #Establecemos una media móvil para 25 días
```

Creamos una columna para cada índice con la media de los últimos 25 días

```python
SP500_data['SMA'] = SP500_data['price'].rolling(window=SMA).mean()   

NASDAQ_data['SMA'] = NASDAQ_data['price'].rolling(window=SMA).mean()

DAX_data['SMA'] = DAX_data['price'].rolling(window=SMA).mean()

IBEX_data['SMA'] = IBEX_data['price'].rolling(window=SMA).mean()
```

Creamos para cada índice una columna con la desviación, otra sumándola a la media y otra restándola

```python
N=1 #Número de desviaciones estándar para las que se ejecuta el modelo
```

```python
SP500_data['STD'] = N*SP500_data['price'].rolling(window=SMA).std()
SP500_data['SMA+STD'] = SP500_data['SMA'] + SP500_data['STD']
SP500_data['SMA-STD'] = SP500_data['SMA'] - SP500_data['STD']

NASDAQ_data['STD'] = N*NASDAQ_data['price'].rolling(window=SMA).std()
NASDAQ_data['SMA+STD'] = NASDAQ_data['SMA'] + NASDAQ_data['STD']
NASDAQ_data['SMA-STD'] = NASDAQ_data['SMA'] - NASDAQ_data['STD']

DAX_data['STD'] = N*DAX_data['price'].rolling(window=SMA).std()
DAX_data['SMA+STD'] = DAX_data['SMA'] + DAX_data['STD']
DAX_data['SMA-STD'] = DAX_data['SMA'] - DAX_data['STD']

IBEX_data['STD'] = N*IBEX_data['price'].rolling(window=SMA).std()
IBEX_data['SMA+STD'] = IBEX_data['SMA'] + IBEX_data['STD']
IBEX_data['SMA-STD'] = IBEX_data['SMA'] - IBEX_data['STD']
```

Representación para SP500

```python
plt.style.use('seaborn')
SP500_data[['price','SMA+STD','SMA-STD']].plot(figsize=(15,8))
```

Representación Nasdaq

```python
plt.style.use('seaborn')
NASDAQ_data[['price','SMA+STD','SMA-STD']].plot(figsize=(15,8))
```

Representación DAX

```python
plt.style.use('seaborn')
DAX_data[['price','SMA+STD','SMA-STD']].plot(figsize=(15,8))
```

Representación IBEX

```python
plt.style.use('seaborn')
IBEX_data[['price','SMA+STD','SMA-STD']].plot(figsize=(15,8))
```

```python
SP500_data['orden'] = np.where(SP500_data['price'] > SP500_data['SMA+STD'],-1,0)
SP500_data['orden'] = np.where(SP500_data['price'] < SP500_data['SMA-STD'],1,SP500_data['orden'])
SP500_data['orden'] = SP500_data['orden'].fillna(0)

NASDAQ_data['orden'] = np.where(NASDAQ_data['price'] > NASDAQ_data['SMA+STD'],-1,0)
NASDAQ_data['orden'] = np.where(NASDAQ_data['price'] < NASDAQ_data['SMA-STD'],1,NASDAQ_data['orden'])
NASDAQ_data['orden'] = NASDAQ_data['orden'].fillna(0)

DAX_data['orden'] = np.where(DAX_data['price'] > DAX_data['SMA+STD'],-1,0)
DAX_data['orden'] = np.where(DAX_data['price'] < DAX_data['SMA-STD'],1,DAX_data['orden'])
DAX_data['orden'] = DAX_data['orden'].fillna(0)

IBEX_data['orden'] = np.where(IBEX_data['price'] > IBEX_data['SMA+STD'],-1,0)
IBEX_data['orden'] = np.where(IBEX_data['price'] < IBEX_data['SMA-STD'],1,IBEX_data['orden'])
IBEX_data['orden'] = IBEX_data['orden'].fillna(0)
```

Para el SP500 tenemos los siguientes resultados

```python
SP500_data['returns'] = SP500_data['price']/SP500_data['price'].shift(1)
SP500_data['Test'] = SP500_data['returns'] ** SP500_data['orden'].shift(1)
SP500_data[['returns', 'Test']].dropna().cumprod().plot(figsize=(15,8))
plt.title("Rendimiento Reversión a la Media Diaria SP500")
```

La estrategia es rentable pero es mejor comprar directamente el índice

Para el Nasdaq tenemos

```python
NASDAQ_data['returns'] = NASDAQ_data['price']/NASDAQ_data['price'].shift(1)
NASDAQ_data['Test'] = NASDAQ_data['returns'] ** NASDAQ_data['orden'].shift(1)
NASDAQ_data[['returns', 'Test']].dropna().cumprod().plot(figsize=(15,8))
plt.title("Rendimiento Reversión a la Media Diaria NASDAQ")
```

La estrategia no es rentable en en el Nasdaq

Para el Dax tenemos

```python
DAX_data['returns'] = DAX_data['price']/DAX_data['price'].shift(1)
DAX_data['Test'] = DAX_data['returns'] ** DAX_data['orden'].shift(1)
DAX_data[['returns', 'Test']].dropna().cumprod().plot(figsize=(15,8))
plt.title("Rendimiento Reversión a la Media Diaria DAX")
```

```python
IBEX_data['returns'] = IBEX_data['price']/IBEX_data['price'].shift(1)
IBEX_data['Test'] = IBEX_data['returns'] ** IBEX_data['orden'].shift(1)
IBEX_data[['returns', 'Test']].dropna().cumprod().plot(figsize=(15,8))
plt.title("Rendimiento Reversión a la Media Diaria IBEX")
```

En este caso parece demasiado caótico, no parece una apuesta segura tampoco.
